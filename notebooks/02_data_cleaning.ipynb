# Melbourne Housing Market - Data Cleaning

This notebook cleans the Melbourne housing dataset by handling missing values and removing incomplete records.

```python
import pandas as pd
import numpy as np
```

```python
# Load the dataset
df = pd.read_csv('../data/Melbourne_housing_FULL.csv')
print(f"Original dataset shape: {df.shape}")
print(f"Original null counts:\n{df.isnull().sum()}")
```

## Data Cleaning Steps

```python
# 1. Drop rows where Price is null
df = df.dropna(subset=['Price'])
print(f"After dropping null Price: {df.shape}")
```

```python
# 2. Drop BuildingArea and YearBuilt columns entirely
df = df.drop(columns=['BuildingArea', 'YearBuilt'])
print(f"After dropping BuildingArea and YearBuilt: {df.shape}")
```

```python
# 3. Fill null values in Car, Bathroom, Bedroom2 with their column median
for col in ['Car', 'Bathroom', 'Bedroom2']:
    median_value = df[col].median()
    df[col] = df[col].fillna(median_value)
    print(f"Filled {col} nulls with median: {median_value}")
```

```python
# 4. Drop rows missing Distance, Postcode, CouncilArea, Regionname, or Propertycount
required_cols = ['Distance', 'Postcode', 'CouncilArea', 'Regionname', 'Propertycount']
df = df.dropna(subset=required_cols)
print(f"After dropping rows with missing required columns: {df.shape}")
```

## Verification

```python
# Print shape before and after cleaning
print("\n=== CLEANING SUMMARY ===")
print(f"Original shape: {pd.read_csv('../data/Melbourne_housing_FULL.csv').shape}")
print(f"Cleaned shape: {df.shape}")
print(f"Rows removed: {pd.read_csv('../data/Melbourne_housing_FULL.csv').shape[0] - df.shape[0]}")
print(f"Columns removed: {pd.read_csv('../data/Melbourne_housing_FULL.csv').shape[1] - df.shape[1]}")
```

```python
# Confirm no nulls remain except Lattitude/Longtitude
null_counts = df.isnull().sum()
print("\nNull value counts after cleaning:")
print(null_counts[null_counts > 0])

# Verify only Lattitude/Longtitude have nulls
cols_with_nulls = null_counts[null_counts > 0].index.tolist()
expected_nulls = ['Lattitude', 'Longtitude']
if set(cols_with_nulls) == set(expected_nulls):
    print("\n✓ Verification passed: Only Lattitude/Longtitude have null values")
else:
    print(f"\n⚠ Warning: Unexpected nulls found in: {cols_with_nulls}")
```

## Save Cleaned Dataset

```python
# Save cleaned dataframe
df.to_csv('../data/melbourne_cleaned.csv', index=False)
print("\nCleaned dataset saved to: data/melbourne_cleaned.csv")
```
